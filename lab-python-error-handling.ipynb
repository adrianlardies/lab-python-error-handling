````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    
    return inventory
````

````python
def calculate_total_price(customer_orders):
    total_price = 0
    for producto in customer_orders:
        while True:
            try:
                precio = float(input(f'¿Cuál es el precio de {producto}? '))
                if precio < 0:
                    print("Error: El precio no puede ser negativo. Inténtalo de nuevo.")
                else:
                    total_price += precio
                    break
            except ValueError:
                print(f'Error: El valor introducido no es un número válido. Inténtalo de nuevo.')
    
    return total_price
````

````python
def get_customer_orders(inventory):
    customer_orders = set()

    while True:
        try:
            num_orders = int(input('¿Cuántos productos deseas añadir? '))
            if num_orders <= 0:
                print('Error: El número de productos debe ser un valor positivo.')
                continue
        except ValueError:
            print('Error: Por favor, introduce un número válido.')
            continue
        
        for _ in range(num_orders):
            while True:
                try:
                    producto = input('Añade a la lista un producto que el cliente desea: ')
                    if producto not in inventory:
                        print(f'Error: El producto "{producto}" no está en el inventario.')
                    elif inventory[producto] <= 0:
                        print(f'Error: El producto "{producto}" no tiene stock disponible.')
                    else:
                        customer_orders.add(producto)
                        break
                except Exception as e:
                    print(f'Ha ocurrido un error al añadir el producto: {e}')
        
        while True:
            try:
                continuar = input('¿Quieres añadir más productos? (S/N): ').upper()
                if continuar == 'S':
                    break
                elif continuar == 'N':
                    return customer_orders
                else:
                    print('Disculpa, no he entendido tu respuesta. Por favor, introduce S o N.')
            except Exception as e:
                print(f'Ha ocurrido un error al procesar tu respuesta: {e}')
````

````python
def update_inventory(customer_orders, inventory):
    inventory_update = {producto: (cantidad - 1 if producto in customer_orders else cantidad) for producto, cantidad in inventory.items()}
    inventory_final = {producto: cantidad for producto, cantidad in inventory_update.items() if cantidad > 0}
    
    return inventory_final
````

````python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique_products = total_products_ordered / len(products) * 100
    order_statistics = (total_products_ordered, percentage_unique_products)
    
    return order_statistics
````

````python
def print_order_statistics(order_statistics):
    print("Estadísticas del pedido:")
    print(f"Total de productos pedidos: {order_statistics[0]}")
    print(f"Porcentaje de productos únicos pedidos: {order_statistics[1]:.2f}%")
````

````python
def print_updated_inventory(inventory_final):
     print("\n".join([f"{producto}: {cantidad}" for producto, cantidad in inventory_final.items()]))
````

````python
def main():
    # Lista de productos iniciales
    products = ['libros', 'peliculas', 'televisores', 'impresoras']
    
    # Paso 1: Inicializar el inventario
    inventory = initialize_inventory(products)
    
    # Paso 2: Obtener pedidos del cliente
    customer_orders = get_customer_orders(inventory)
    
    # Paso 3: Actualizar el inventario
    inventory = update_inventory(customer_orders, inventory)
    
    # Paso 4: Calcular estadísticas del pedido
    order_statistics = calculate_order_statistics(customer_orders, products)
    
    # Paso 5: Imprimir estadísticas del pedido
    print_order_statistics(order_statistics)
    
    # Paso 6: Imprimir inventario actualizado
    print_updated_inventory(update_inventory(customer_orders, inventory))

    # Paso 7: Imprimir precio total del pedido
    print(calculate_total_price(get_customer_orders()))

# Llamada a la función principal
if __name__ == "__main__":
    main()
````
